```python
# Do the number of JobRole affect the employees WorkLifeBalance?
sns.boxplot(x='JobRole', y='WorkLifeBalance', data=df, palette='viridis')
plt.title('Job Role vs Work Life Balance')
plt.xticks(rotation=45)
plt.show()
```

Looking at this **boxplot (JobRole vs WorkLifeBalance):**

---

### Observations:

1. **Median values**:

   * Across all job roles, the median **WorkLifeBalance** hovers around **2â€“3**.
   * This suggests no job role shows a drastically higher or lower balance compared to others.

2. **Spread (IQR)**:

   * For most roles (Sales Executive, Research Scientist, Laboratory Technician, Manufacturing Director, Healthcare Representative, Manager), the spread is similar.
   * The distribution is fairly balanced and consistent.

3. **Outliers**:

   * A few roles (Sales Representative, Research Director, Human Resources) show many outliers at **1** and **4**, but they donâ€™t shift the overall median much.
   * These outliers indicate some individuals in those roles experience extreme work-life balance (either very poor or very good).

4. **Overall trend**:

   * No clear distinction between roles â†’ **job role does not significantly affect work-life balance** in this dataset.
   * Variability within roles is larger than variability across roles.

---

âœ… **Interpretation:**
Work-life balance seems **consistent across different job roles**, meaning factors other than role designation (like overtime, workload, or company policies) might be more important drivers.

```python
# Does the JobInvolvement affect the Relationship Satisfaction?
sns.countplot(x='JobInvolvement', hue='RelationshipSatisfaction', data=df, palette='viridis')
plt.title('Job Involvement vs Relationship Satisfaction')
plt.show()
```

Looking at this **countplot (JobInvolvement vs RelationshipSatisfaction):**

---

### Observations:

1. **Most employees have JobInvolvement = 3**

   * This is the dominant category, with the highest count across all satisfaction levels.
   * Within this group, **RelationshipSatisfaction = 3 and 4** are the most common.

2. **JobInvolvement = 2**

   * The second-largest group.
   * RelationshipSatisfaction is spread across 2, 3, and 4, but 3 is still higher.

3. **JobInvolvement = 4**

   * Smaller group compared to 2 and 3.
   * RelationshipSatisfaction is still more skewed toward 3 and 4.

4. **JobInvolvement = 1 (very low involvement)**

   * Very few employees fall here.
   * Satisfaction is also low in count but relatively balanced across 1â€“4.

---

### Key Insights:

* **Employees with moderate to high job involvement (3 & 4) tend to report higher relationship satisfaction (3 or 4).**
* **Low involvement (1 & 2) employees are more mixed in their satisfaction**, some are satisfied but many are in the mid-range.
* **Overall trend** â†’ Higher JobInvolvement is **positively associated** with higher RelationshipSatisfaction.

---

âœ… **Interpretation:**
It looks like employees who are **more involved in their jobs tend to be more satisfied in their workplace relationships**. This could mean engagement in work also improves interpersonal satisfaction at work.
